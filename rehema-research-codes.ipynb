```python
# importing important libraries
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
from scipy.integrate import odeint
import matplotlib.pyplot as plt
```

## Exponential growth

```python
# Initialization
u=10
r=0.45
S=[]
K=np.linspace(0,10,100)
#Creating the list of exponential funcion
for i in range(len(K)):
    v=u*np.exp(r*K[i])
    S.append(v)
#Ploting the curve
fig = plt.figure(figsize=(5,4))
plt.plot(K,S,linewidth=3.0)
plt.xlabel("Time (t)")
plt.ylabel("Population size")
plt.show()
```

## Logistic growth

```python
# function that returns dydt
def model(y,t):
    k =80
    r=0.5
    dydt =r*y*((k-y)/k)
    return dydt

y0 =20                 # initial condition
t= np.linspace(0,20)   # time range point
y = odeint(model,y0,t) # solve ODE
# plot results
fig = plt.figure(figsize=(5,4))
plt.plot(t,y,linewidth=3.0)
plt.xlabel('Time')
plt.ylabel('Population size')
plt.show()
```

## Sensitivity Analysis

```python
# Initialization
A=[]
B=[]
C=[]
D=[]
T=[]
m=0.4    # maturation rate
d_1=0.3  # juvenile death rate
d_2=0.35 # adult death rate
f=0.5    # adult fertilty rate
# Define the range point of K with 10% interval
K=np.linspace(0.4,0.44,2)
# Function that return effect of 10% increase in each parameter variable
for i in range(len(K)):
    lamda=(2-K[i]-d_2-m+((2-K[i]-d_2-m)**2-4*((1-K[i]-m)*(1-d_2)-f*m))**0.5)/2  #d_1
    lamda1=(2-d_1-K[i]-m+((2-d_1-K[i]-m)**2-4*((1-d_1-m)*(1-K[i])-f*m))**0.5)/2 #d_2
    lamda2=(2-d_1-d_2-K[i]+((2-d_1-d_2-K[i])**2-4*((1-d_1-K[i])*(1-d_2)-f*K[i]))**0.5)/2#m
    lamda3=(2-d_1-d_2-m+((2-d_1-d_2-m)**2-4*((1-d_1-m)*(1-d_2)-K[i]*m))**0.5)/2 #f
    A.append(lamda)
    B.append(lamda1)
    C.append(lamda2)
    D.append(lamda3)
T.append(A[1]-A[0]) 
T.append(B[1]-B[0])
T.append(C[1]-C[0])
T.append(D[1]-D[0])
plotdata = pd.DataFrame(
    {"pies": [vals for vals in T]}, 
    index=["$d_1$", "$d_2$", "$m$", "$f$"])
#Plot a bar chart
plotdata.plot(kind="bar")
plt.ylabel("Population growth $(\lambda)$")
plt.show()
```

## Roots of polynomial when $c<c^*$

```python
A1=[]
A2=[]
c=0.08    # wave speed
D_1=0.3   # juvenile dispersal rates 
D_2=0.3   # adult dispersal rates
t=0.6     # maturation rate
d_1=0.1   # mortality of juvenile
d_2=0.4   # mortality rate of adult
b=0.65    # reproduction rate
F=np.linspace(-2,2.3,10000)
for i in range(len(F)):
    P_c=(d_1+t+F[i]*(c-D_1*F[i]))*(d_2+F[i]*(c-D_2*F[i]))-b*t
    V=0
    A2.append(V)
    A1.append(P_c)
#ploting 
plt.figure(figsize=(6,5))
plt.plot(F,A1,"b--",label="P($\lambda$)")
plt.plot(F,A2,"r-.",label="$\lambda$")
plt.legend()
plt.title("$c<c^*$")
plt.xlabel("$\lambda$")
plt.ylabel("$P(\lambda)$")
plt.show()
```
